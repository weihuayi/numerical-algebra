***
THEOREM 2.1. Let A be nonsingular.Then
$$
\min \left\{ 
    \frac {\| \delta \boldsymbol A \| _2} {\| \boldsymbol A \| _2}:
    \boldsymbol A + \delta \boldsymbol A\ singular
\right\}
=
\frac {1} {\| \boldsymbol A^{-1} \| _2 \cdot \| \boldsymbol A \| _2}
=
\frac {1} {\kappa (\boldsymbol A)}
$$
***
### 特征关系
在考虑这个问题之前,我需要先展示几个和 $ \boldsymbol A^* \boldsymbol A $ 有关的性质,对 $ \boldsymbol A^* \boldsymbol A $ 做酉分解 :
$$ 
\boldsymbol A^* \boldsymbol A 
=
\boldsymbol U \boldsymbol \Lambda \boldsymbol U^*
$$
则:
$$
\begin{alignat}{2}
\boldsymbol A \boldsymbol A^* 
&= 
\boldsymbol A (\boldsymbol A^* \boldsymbol A) \boldsymbol A^{-1}
=
\boldsymbol A (\boldsymbol U \boldsymbol \Delta \boldsymbol U^*) \boldsymbol A^{-1}
=
(\boldsymbol A \boldsymbol U) \boldsymbol \Lambda (\boldsymbol A \boldsymbol U)^{-1} \\
\boldsymbol A^{-1*} \boldsymbol A^{-1} 
&= 
(\boldsymbol A \boldsymbol A^*)^{-1}
=
(\boldsymbol A \boldsymbol U) \boldsymbol \Lambda^{-1} (\boldsymbol A \boldsymbol U)^{-1} \\
\end{alignat}
$$
反复应用这些等式就可以大致理解书上的证明（但是书上写的有误，是$ \boldsymbol y^* $ 不是 $ \boldsymbol y^T $ ）.不过我接下来要抛弃书上的构造性的证明，从原理出发,论述找 $ \delta \boldsymbol A $ 的原理和构造 $ \delta \boldsymbol A $ 的原理.
***
### 问题分析 
$$$$
对 $ \boldsymbol A \boldsymbol A^* $ 做酉分解,则对角矩阵 $ \boldsymbol \Lambda $ 可以写作:
$$
\boldsymbol \Lambda
=
\begin{bmatrix}
\left| \lambda_0 \right| ^2 & \quad  &  \quad \\
\quad              & \ddots &  \quad \\
\quad              & \quad  &  \left| \lambda _{n - 1} \right| ^2
\end{bmatrix},
\left| \lambda_0 \right| \ge \cdots \ge \left| \lambda _{n - 1} \right|
$$
酉阵 $ \boldsymbol U $ 可以写作:
$$
\boldsymbol U
=
\begin{bmatrix}
\boldsymbol u_0 & \cdots & \boldsymbol u_{n - 1}
\end{bmatrix}
$$
那么:
$$
\boldsymbol A^* \boldsymbol A
=
\boldsymbol U \boldsymbol \Lambda \boldsymbol U^*
=
\boldsymbol U^*
\begin{bmatrix}
\left| \lambda_0 \right| ^2 & \quad  & \quad \\
\quad              & \ddots & \quad \\
\quad              & \quad  & \left| \lambda _{n - 1} \right| ^2
\end{bmatrix}
\boldsymbol U
$$
对 $ \boldsymbol A $ 做特征范数下最小程度的扰动,就应当是使得 $ (\boldsymbol A + \delta \boldsymbol A)^* (\boldsymbol A + \delta \boldsymbol A) $ 具有形式:
$$
(\boldsymbol A + \delta \boldsymbol A)^* (\boldsymbol A + \delta \boldsymbol A)
=
\boldsymbol U^*
\begin{bmatrix}
\left| \lambda_0 \right| ^2 & \quad  & \quad            & 0      \\
\quad              & \ddots & \quad                     & \vdots \\
\quad              & \quad  & \left| \lambda _{n - 2} \right| ^2 & 0 \\
0                  & \cdots & 0                         & 0
\end{bmatrix}
\boldsymbol U
$$
在这个等式中,并不一定需要$ \boldsymbol U $ 是等于$ \boldsymbol A^* \boldsymbol A $分解中的$ \boldsymbol U $,当然如果是相等的$ \boldsymbol U $，那是最好的.并且在这里我们基本可以猜出结论 $ \| \delta \boldsymbol A \|_2 = \left| \lambda_{n - 1} \right| = \| \boldsymbol A^{-1} \|_2^{-1} $ ,对$ \boldsymbol A^* \boldsymbol A $ 而言:
$$
\boldsymbol A^* \boldsymbol A
\begin{bmatrix}
\boldsymbol u_0 & \cdots & \boldsymbol u_{n - 1}
\end{bmatrix}
=
\begin{bmatrix}
\left| \lambda_0 \right| ^2 \boldsymbol u_0 & \cdots & \left| \lambda_{n - 1} \right| ^2 \boldsymbol u_{n - 1}
\end{bmatrix}
$$
对$ (\boldsymbol A + \delta \boldsymbol A)^* (\boldsymbol A + \delta \boldsymbol A) $ 而言:
$$
(\boldsymbol A + \delta \boldsymbol A)^* (\boldsymbol A + \delta \boldsymbol A)
\begin{bmatrix}
\boldsymbol u_0 & \cdots & \boldsymbol u_{n - 1}
\end{bmatrix}
=
\begin{bmatrix}
\left| \lambda_0 \right| ^2 \boldsymbol u_0 & \cdots & \left| \lambda_{n - 2} \right| ^2 \boldsymbol u_{n - 2} & 0
\end{bmatrix}
$$
所以我们的目标就是为了构造这样的 $ \delta \boldsymbol A $ 使得:
$$
(\boldsymbol A + \delta \boldsymbol A)^* (\boldsymbol A + \delta \boldsymbol A) \boldsymbol u_{n - 1} = \boldsymbol 0
$$
但是这样构造 $ \delta \boldsymbol A $ 是十分困难的,因为 $ (\boldsymbol A + \delta \boldsymbol A)^* (\boldsymbol A + \delta \boldsymbol A) $ 关于未知量 $ \delta \boldsymbol A $ 是一个多项式结构,这非常不利于计算未知量 $ \delta \boldsymbol A $ ,一般来说连乘形式的方程比较更容易解，所以将$ (\boldsymbol A + \delta \boldsymbol A)^* (\boldsymbol A + \delta \boldsymbol A) $写作:
$$
(\boldsymbol A + \delta \boldsymbol A)^* (\boldsymbol A + \delta \boldsymbol A)
=
(\boldsymbol A + \boldsymbol A_{1 + \delta})^* (\boldsymbol A + \boldsymbol A_{1 + \delta})
=
\boldsymbol A_{1 + \delta}^* \boldsymbol A^* \boldsymbol A \boldsymbol A_{1 + \delta}
$$
现在只要构造满足以下条件的 $ \boldsymbol A_{1 + \delta} $ 就可以了:
$$
\boldsymbol A_{1 + \delta}^* \boldsymbol A^* \boldsymbol A \boldsymbol A_{1 + \delta} \boldsymbol u_{n - 1} 
= 
(\boldsymbol A + \delta \boldsymbol A)^* (\boldsymbol A + \delta \boldsymbol A) \boldsymbol u_{n - 1} 
=
\boldsymbol 0
$$
***
### 解答过程
通过上面的分析，我们知道选择$ \boldsymbol u_{u - 1} $作为$ \boldsymbol A + \delta \boldsymbol A $的特征值为0的特征向量是合理的,考虑到:
$$
\begin{alignat}{2}
\boldsymbol A^* \boldsymbol A \boldsymbol u_{u - 1} 
&= 
\| \lambda_{n - 1} \|^2 \boldsymbol u_{u - 1} 
=
\| \lambda_{n - 1} \|^2 \boldsymbol u_{u - 1} \boldsymbol u_{u - 1}^* \boldsymbol u_{u - 1}
\end{alignat}
$$
所以有:
$$
\boldsymbol A^* \boldsymbol A ( \boldsymbol I - \boldsymbol u_{u - 1} \boldsymbol u_{u - 1}^*) \boldsymbol u_{u - 1} = \boldsymbol 0
$$ 
这说明当选取$ \delta \boldsymbol A = -\boldsymbol A \boldsymbol u_{u - 1} \boldsymbol u_{u - 1}^* $时,$ \boldsymbol u_{u - 1} $确实是$ \boldsymbol A + \delta \boldsymbol A $的特征值为0的特征向量,下面反复应用前面"特征关系"证明$ \| \delta \boldsymbol A \|_2 $确实是$ \| \cdot \|_2 $是最小的:
$$
\mathrm {rank} (\delta \boldsymbol A^* \delta \boldsymbol A) 
= 
\mathrm {rank} (-\boldsymbol A \boldsymbol u_{u - 1} \boldsymbol u_{u - 1}^* )
=
\mathrm {rank} ( \boldsymbol u_{u - 1} \boldsymbol u_{u - 1}^* )
=
1
$$
说明 $ \delta \boldsymbol A^* \delta \boldsymbol A $ 只有一个特征值不为0 的特征向量:
$$
\begin{alignat}{2}
\delta \boldsymbol A^* \delta \boldsymbol A \boldsymbol u_{n - 1}
&=
(\boldsymbol A \boldsymbol u_{n - 1} \boldsymbol u_{n - 1}^*)^* (\boldsymbol A \boldsymbol u_{n - 1} \boldsymbol u_{n - 1}^*) \boldsymbol u_{n - 1} \\
&=
\boldsymbol u_{n - 1} \boldsymbol u_{n - 1}^* \boldsymbol A^* \boldsymbol A \boldsymbol u_{n - 1} \\
&=
\left| \lambda_{n - 1} \right|^2 \boldsymbol u_{n - 1} 
\end{alignat}
$$
所以$ \delta \boldsymbol A^* \delta \boldsymbol A $具有一个的特征值 $ \left| \lambda_{n - 1} \right|^2 = \| \boldsymbol A^{-1} \|_2^{-2} $ 并且其他特征值都为0.至此全部证毕,并且:
$$
\delta \boldsymbol A = -\boldsymbol A \boldsymbol u_{u - 1} \boldsymbol u_{u - 1}^*,
\| \delta \boldsymbol A \|_2 = \left| \lambda_{n - 1} \right|
$$
